```python
import easyocr

def run_text_ocr(image_path):
    print("Initializing OCR Engine (loading English models)...")
    # This downloads the pre-trained weights on the first run
    reader = easyocr.Reader(['en'])
    
    print(f"Processing image: {image_path}\n")
    # Extract text locations, characters, and confidence scores
    results = reader.readtext(image_path)
    
    print("=" * 40)
    print("          DETECTION OUTPUT          ")
    print("=" * 40)
    
    if not results:
        print("No text elements detected in the image.")
        return
        
    for index, (bbox, text, confidence) in enumerate(results, 1):
        print(f"[{index}] Text Found: '{text}'")
        print(f"    Confidence: {confidence * 100:.2f}%\n")
    print("=" * 40)

if __name__ == "__main__":
    # Make sure 'sample_text.png' is saved in this exact same folder!
    sample_image = "R.jfif"
    
    try:
        run_text_ocr(sample_image)
    except FileNotFoundError:
        print(f"Error: Could not find '{sample_image}'. Please check the file name and path.")
```

```python
import torch
import torchvision.transforms as transforms
from torchvision.models import mobilenet_v3_small, MobileNet_V3_Small_Weights
from PIL import Image
import urllib.request
import json

def fetch_imagenet_labels():
    """Downloads human-readable names for the 1000 ImageNet categories."""
    url = "https://raw.githubusercontent.com/AnishShah/imagenet-labels-json/master/imagenet-labels.json"
    try:
        with urllib.request.urlopen(url) as response:
            return json.loads(response.read().decode())
    except Exception:
        print("Network error: Using fallback labels...")
        # Fallback generic mapping if internet drops
        return {str(i): f"Class ID {i}" for i in range(1000)}

def classify_image(image_path):
    print("Loading pre-trained MobileNetV3 deep learning architecture...")
    # Load the best available weights for MobileNetV3 Small
    weights = MobileNet_V3_Small_Weights.DEFAULT
    model = mobilenet_v3_small(weights=weights)
    model.eval()  # Put model in evaluation (inference) mode
    
    # 1. Image Preprocessing Pipelines
    # CNN models expect uniform tensor dimensions, colors normalized, and cropped shapes
    preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    
    # 2. Load and transform the target image
    raw_img = Image.open(image_path).convert('RGB')
    input_tensor = preprocess(raw_img)
    input_batch = input_tensor.unsqueeze(0)  # Expand dimensions to create a batch of 1 image
    
    print(f"Analyzing image structure: '{image_path}'...\n")
    
    # 3. Model Inference (Forward Pass)
    with torch.no_grad():
        raw_outputs = model(input_batch)
    
    # 4. Convert raw model outputs (logits) to relative percentage scores
    probabilities = torch.nn.functional.softmax(raw_outputs[0], dim=0)
    
    # Extract the highest probability category
    top_probability, top_class_id = torch.topk(probabilities, 1)
    
    # Map the winning class index number to its readable label string
    labels = fetch_imagenet_labels()
    object_label = labels.get(str(top_class_id.item()), "Unknown Object")
    
    print("=" * 40)
    print("        CLASSIFICATION OUTPUT        ")
    print("=" * 40)
    print(f"Predicted Label:   {object_label.title()}")
    print(f"Confidence Score:  {top_probability.item() * 100:.2f}%")
    print("=" * 40)

if __name__ == "__main__":
    sample_img_file = "OIP.webp"
    
    try:
        classify_image(sample_img_file)
    except FileNotFoundError:
        print(f"Error: Could not locate '{sample_img_file}'. Please place an image in this directory.")
```

```python
%%writefile app.py
import streamlit as st
import easyocr
import torch
import torchvision.transforms as transforms
from torchvision.models import mobilenet_v3_small, MobileNet_V3_Small_Weights
from PIL import Image
import numpy as np

# --- Page Setup ---
st.set_page_config(page_title="AI Vision Toolkit", page_icon="📸", layout="centered")

st.title("📸 AI Vision Toolkit")
st.markdown("### Project 4: Image & Text Recognition System")
st.write("Upload an image, pick your AI task from the sidebar, and run analysis instantly!")

# --- Caching Engines for Performance ---
@st.cache_resource
def get_ocr_reader():
    return easyocr.Reader(['en'])

@st.cache_resource
def get_mobilenet_model():
    weights = MobileNet_V3_Small_Weights.DEFAULT
    model = mobilenet_v3_small(weights=weights)
    model.eval()
    categories = weights.meta["categories"]
    return model, categories

# --- Sidebar Controls ---
st.sidebar.header("🔧 Task Configuration")
selected_task = st.sidebar.radio(
    "Select AI Mode:",
    ("Text Recognition (OCR)", "Image Classification")
)

# --- Main App Interface ---
uploaded_file = st.file_uploader("Drop your target image file here...", type=["png", "jpg", "jpeg", "webp"])

if uploaded_file is not None:
    img = Image.open(uploaded_file).convert('RGB')
    # CHANGED: use_column_width=True for compatibility with Streamlit v1.30.0
    st.image(img, caption="Target Input View", use_column_width=True)
    
    if st.button("🚀 Execute Neural Network"):
        
        # --- TASK 1: TEXT RECOGNITION ---
        if selected_task == "Text Recognition (OCR)":
            with st.spinner("Extracting alphanumeric data..."):
                reader = get_ocr_reader()
                img_np = np.array(img)
                ocr_results = reader.readtext(img_np)
                
            st.subheader("📝 Text Extraction Results")
            if not ocr_results:
                st.warning("No readable text structures detected.")
            else:
                for idx, (bbox, extracted_text, confidence) in enumerate(ocr_results, 1):
                    st.success(f"**[{idx}] Text:** `{extracted_text}`  \n*Confidence Score:* {confidence * 100:.1f}%")

        # --- TASK 2: IMAGE CLASSIFICATION ---
        elif selected_task == "Image Classification":
            with st.spinner("Running CNN Feature Extraction..."):
                model, categories = get_mobilenet_model()
                
                preprocess = transforms.Compose([
                    transforms.Resize(256),
                    transforms.CenterCrop(224),
                    transforms.ToTensor(),
                    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                ])
                
                tensor_input = preprocess(img).unsqueeze(0)
                
                with torch.no_grad():
                    logits = model(tensor_input)
                
                probs = torch.nn.functional.softmax(logits[0], dim=0)
                top_prob, top_class_id = torch.topk(probs, 1)
                predicted_name = categories[top_class_id.item()]
                
            st.subheader("🔍 Computer Vision Classification")
            st.metric(label="Predicted Object Label", value=predicted_name.title())
            st.metric(label="Prediction Confidence", value=f"{top_prob.item() * 100:.2f}%")
else:
    st.info("Awaiting input data. Please drop an image file to engage processing engines.")
```
